#Machine learing section

```python
# Separate features (X) and target (y)
x = data.drop('Salary', axis=1)  # Features: all columns except 'Salary'
y = data['Salary']             # Target: the 'Salary' column
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Set a random seed for reproducibility
# This ensures that the random split will be the same every time you run the code
np.random.seed(42)

# Split the data into training and testing sets
# test_size=0.2 means 20% of the data will be used for testing
# The remaining 80% will be used for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

dealing with categorical variables

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Create a ColumnTransformer object
# This object will handle the transformation of specific columns in our dataset
ct = ColumnTransformer(
    [('encoder', OneHotEncoder(), [1, 2])],  # Apply OneHotEncoder to columns 1 and 2 (Education Level and Job Role)
    remainder='passthrough'  # Keep the remaining columns as they are
)

# Fit the ColumnTransformer to the training data and transform it
# This step learns the categories from the training data and applies OneHotEncoding
x_train = ct.fit_transform(x_train)

# Transform the test data using the fitted ColumnTransformer
# This ensures that the test data is transformed in the same way as the training data
x_test = ct.transform(x_test)
```

Fit linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

evaluate the model

```python
model.score(x_test,y_test)
```

```python
plt.scatter(x_test[:, -1].toarray(), y_test, color='coral', edgecolors='black', alpha=0.7, label='Test Data')
plt.plot(x_test[:, -1].toarray(), y_pred, color='blue', linewidth=1, label='Regression Line')


plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience (Test Data)')

plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error

mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

```

```python
eval_df = pd.DataFrame([[mae, mse, rmse, r2]], columns=['MAE', 'MSE', 'RMSE', 'R2 Score']).round(2)
```

```python
eval_df # model is performing very well, with a high R² score indicating strong predictive power.
```
